#### "data" değişkenindeki veri setinize tork adında yeni bir sütun/değişken ekleyin. Bu sütunu aşağıdaki formüle göre hesaplayıp oluşturun.


**Tork (Torque)**: 

Tork, motorun dönme kuvvetinin; araçta tekerlekleri itme kuvvetine dönüştürülmesini sağlamaktadır. Yani bir nevi, aracın çekiş gücü ile ilgilidir. 

Dolayısı ile Tork'u yüksek olan arabaların daha pahalı olması beklenebilir.

**Torque = 5252 x HP(horsepower) / RPM(peakrpm)**


Öncelikle formülü yazıp torque değişkenine atayın.

Daha sonra oluşturduğunuz bu değişkeni veri setinize sütun olarak ekleyin.

```python
torque = 5252 * data['horsepower'] / data['peakrpm']
data['torque'] = 5252 * data['horsepower'] / data['peakrpm']
```

```python
data['torque'] = pd.Series(torque, index=data.index)
```

#### "data" değişkenindeki veri setinizdeki Torque ile Fiyat arasındaki ilişkiyi grafikle (regplot ile) gösterin.

```python
sns.regplot(x='torque', y='price', data=data)
plt.show()
```

#### Torque değişkeninin dağılımını grafikle (distplot ile) gösterin.

```python
sns.distplot(data['torque'])
plt.show()
```

#### data'daki verilerinizi cars değişkenine kopyalayarak atayın.

#### Kategorik değişkenler için dummy variable'ları alın, one hot dönüşümünü gerçekleştirin. 2 tane sınıfı olan değişkenlerden birini veri setinden silin.

**pd.get_dummies()**
dummies_list = ['fueltype', 'aspiration', 'carbody','drivewheel',
                'enginelocation', 'enginetype', 'cylindernumber']

```python
cars = data.copy()

dummies_list = ['fueltype', 'aspiration', 'carbody','drivewheel', 'enginelocation', 'enginetype', 'cylindernumber']

for var in dummies_list:
    dummies = pd.get_dummies(cars[var], prefix=var)
    cars = pd.concat([cars, dummies], axis=1)
    cars.drop([var], axis=1, inplace=True)
```

```python
cars.drop(['fueltype_gas'], axis=1, inplace=True)
```
